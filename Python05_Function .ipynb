```python
def my_function(*numbers):
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num
print(my_function( 3, 7, 2, 9, 1))
```

```python
def changecase(func):
    def myinner():
        return func().upper()
    return myinner
@changecase
def myfunction():
    return "Hello Sally"
@changecase
def otherfunction():
    return "I am speed!"
print(myfunction())
print(otherfunction())
```

```python
def changecase(func):
    def myinner(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return myinner
@changecase
def myfunction(nam):
    return "Hello " + nam
print(myfunction("John"))
```

```python
def changecase(func):
    def myinner():
        return func().upper()
    return myinner
def addgreeting(func):
    def myinner():
        return "Hello " + func() + " Have a good day!"
    return myinner
@changecase
@addgreeting
def myfunction():
    return "Tobias"
print(myfunction())
```

```python
def fibonacci(n):
    if n <= 1:
         return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(7))
```

```python
def count_up_to(n):
 count = 1
 while count <= n:
    yield count
    count += 1
for num in count_up_to(5):
    print(num)
```

```python
def simple_gen():
 yield 1
 yield 2
gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen)) # This will raise StopIteration
```

```python
def my_gen():
    try:
        yield 1
        yield 2
        yield 3
    finally:
        print ("Generator closed")
gen = my_gen()
print (next(gen))
gen.close()
```
